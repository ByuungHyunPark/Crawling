# Page와 Date를 고려한 뉴스기사 가져오기
- 다음뉴스기사 url의 구성은 다음과 같다. <br>
https://news.daum.net/breakingnews/{1}?page={2}&regDate={3}

    - {1} : 기사 종류 ex)digital, social, sports, politics
    - {2} : 페이지 , 1~50 정도 존재
    - {3} : 날짜,  ex) 20200727

### for문을 통해 , 2020년07월27일 모든 페이지를 불러와보기
- page 가 변하도록 for문 사용

```python
Date = 20200727
#page = 1~ 끝까지.

#url = 'https://news.daum.net/breakingnews/digital?page={}&regDate={}'.format(page, Date)
```

```python
from bs4 import BeautifulSoup

page = 1                 #1페이지부터 시작
max_page = 0
Date = '20200727'

df_list = []
flag = True

while(True):
    
    
    response = requests.get('https://news.daum.net/breakingnews/digital?page={}&regDate={}'\
                            .format(page, Date))
    html = response.text
    soup = BeautifulSoup(html , 'html.parser')
    
    
    ## 페이지가 더 넘어갈곳이 없다면 , 위와 같은 조건문을 거치게됨.
    ## 페이지가 더이상 갈 곳이 없다면 , 아래 조건문의 select문에  '해당 일자에 데이터가 없습니다.' 라는 문구가 생기며
    ## len은 1값을 갖게 되며 , break문을 통해 while을 빠져나올 수 있다.
    if (len(soup.select('#mArticle > div.box_etc > p'))!=0):
        print('마지막 페이지까지 왔으므로 while문 종료')
        break
    
    #if soup.select('#mArticle > div.box_etc > p').text) == '해당 일자에 데이터가 없습니다.':
    #    break
    
    ##mArticle > div.box_etc > ul > li:nth-child(1) > div > strong > a
    for i in soup.select('#mArticle > div.box_etc > ul > li'):
        row_list = []

        news_title = i.select('div > strong > a')[0]
        
        row_list.append(news_title.text)
        row_list.append(news_title.get('href'))
        
        df_list.append(row_list)    
    

    print(page) #진행상황을 알기 위해 출력
    page += 1
```

```python
df_list
```

```python
df = pd.DataFrame(df_list, columns=['기사 제목','url'])
df
```
