# BeautifulSoup
- html 코드를 Python이 이해하는 객체 구조로 변환하는 Parsing을 맡고 있다.

```python
#pip install bs4
import requests
from bs4 import BeautifulSoup

req = requests.get('https://news.v.daum.net/v/20200505000102404')

##HTML소스 가져오기
html = req.text
print(html[:700])
```

##### BeautifulSoup으로 html소스를 python객체로 변환하기
- 첫 인자는 html소스코드, 두 번째 인자는 어떤 parser를 이용할지 명시.
- 주로 python 내장 html.parser를 사용함

```python
soup = BeautifulSoup(html , 'html.parser')
print(soup)
```

### selector를 사용
- Xpath와 다르다는 것이 중요
- 크롤링 하고자 하는것 우클릭->inspect
